### 문자열  
문자들의 **순서가 있는 변경 불가능한** 시퀸스 자료형

```python
print('He \'s a boy.')
print('첫째 줄 \n둘째 줄')
print('탭\t탭')
print('백슬래시\\')
```

**f-string**  
문자열 내에 변수나 표현식의 결과를 손쉽게 삽입하는 방법

```python
name = '임경수'
age = 25
greeting = f'안녕하세요, 이름 {name} 나이{age}'
print(greeting)
```

**f-string 심화**

```python
#자리수 조절
print('--------자리수 조절---------')
number = 123456.123456
print(f'{number:.3f} 소숫점 자리수 조절')
print(f'{number:,.0f} 천 단위 콤마')
print(f'{number:10.2f}')#총 너비10칸 우측 정렬

#패딩 (정렬 및 공백)
print('\n----------패 딩-------------')
print(f'{"우측 정렬 12칸":>12}')
print(f'{"중앙 정렬 12칸":^12}')
print(f'{"좌측 정렬 12칸":<12}')

#진법 변환
print('\n---------진법 변환----------')
number = 255
print(f'16진법: {number:#x}')
print(f'2진법: {number:#b}')
print(f'8진법: {number:#o}')

#날짜 포멧팅
print('\n--------날짜 포맷팅----------')
from datetime import datetime
now = datetime.now()
print(f'{now:%y년 %m월 %d일 %H시 %M분}')#%값 사이사이 자유입력가능
print(f'{now:%Y-%m-%d %H:%M}')#%값 사이사이 자유입력가능
print(f"{now.year}년 {now.month}월 {now.day}일 {now.hour:02d}:{now.minute}:{now.second}")
```

**문자열 슬라이싱**

```python
str = 'hello'
print(str[-3:-6:-1])
print(str[2::-1])
print(str[-5:-3])
print(str[-1:-6:-1])
print(str[::-1])
```

**문자열의 불변성**

**진법 표현하기**

```python
print(0b1111) # 1111_2을 10진수로 표현
print(0o1111) # 1111_8을 10진수로 표현
print(0xfab3) # fab3_16을 10진수로 표현
```

**부동 소수점 오차**  
컴퓨터가 2진수로 계산하고 저장하기 때문에 발생하는문제

```python
#부동 소수점 오차
a = 3.2 - 3.1
b = 1.2 - 1.1
print(a)
print(b)
print(a == b)
#부동 소수점 오차 해결
from decimal import Decimal
a = Decimal('3.2') - Decimal('3.1')
b = Decimal('1.2') - Decimal('1.1')
print(a)
print(b)
print(a == b)
```
