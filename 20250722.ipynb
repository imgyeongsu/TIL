### 리스트
- **모든 종류의 데이터를 담을 수 있음**  
- 변경이 가능하다.

### 튜플
- 변경이 불가능하다.
- 단일 요소 튜플은 반드시 **Trailing comma**를 사용해야한다.
- 여러 요소가 있으면 굳이 안묶어도 튜플로 인식한다.
- 파이썬 내부에서 동작할때 사용된다. 안정성과 무결성이 보장되기 때문

```python
print(type(()))
print(type(('a'))) # 단일 요소
print(type((1))) # 단일 요소
print(type((1,)))
tuple1 = 1,2,3,4,5
print(type(tuple1))
```

### range의 규칙
stop값은 생성되는 시퀀스에 포함되지 않음  
step값은 숫자 시퀀스의 간격과 방향을 결정

```python
print(list(range(2,9,2)))
print(list(range(21,9,-2)))
```

### dict
key-value 쌍으로 이루어진  
순서와 중복이 없는 변경 가능한 자료형  

**key 규칙**  
- 고유해야 함
- 변경 불가능한 자료형만 가능 - list, dict 불가능

**value 규칙**  
어떤 자료형이든 자유롭게 사용할 수 있음

```python
dic = dict({'a' : 1,
            'b' : 2,
            'a' : 3})
print(type(dic.items()))
```

### 세트 set
순서와 중복이 없는 변경 가능한 자료형

```python
a = set([1])
b = {1}
print(type(a), type(b))
print(a, b)
print(a == b)
```

**set의 집합 연산**

```python
set1 = {1, 2, 3}
set2 = {3, 6, 9}

#합집합
print('합집합(|) : ',set1 | set2)
print('차집합(-) : ',set1 - set2)
print('교집합(&) : ',set1 & set2)
```

### is 와 ==
is는 **싱글톤** 객체에만 사용 권장

```python
a =[1,2,3]
b = a.copy()
c = a

print(b is a)
print(c is a)
```
